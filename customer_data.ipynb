```python
def plot_columns_against_churn(df):
    # Separera numeriska och kategoriska kolumner
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

    # LÃ¤gg till 'SeniorCitizen' till kategoriska kolumner om den hanteras som numerisk
    if 'SeniorCitizen' in numerical_cols:
        numerical_cols.remove('SeniorCitizen')
        categorical_cols.append('SeniorCitizen')

    # Ta bort 'Churn' frÃ¥n kategoriska kolumner om den finns dÃ¤r
    if 'Churn' in categorical_cols:
        categorical_cols.remove('Churn')

    # Plotta numeriska kolumner
    for col in numerical_cols:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='Churn', y=col, data=df)
        plt.title(f"{col} vs Churn")
        plt.show()

    # Plotta kategoriska kolumner
    for col in categorical_cols:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=col, hue='Churn', data=df)
        plt.title(f"{col} vs Churn")
        plt.xticks(rotation=45)
        plt.show()

plot_columns_against_churn(train_df)
```

# Finns det data som ej borde anvÃ¤ndas?

Jag tog bort tvÃ¥ kolumner, TotalCharges och gender, frÃ¥n datan nÃ¤r jag fÃ¶rberedde den fÃ¶r modellen. Anledningen till att 'TotalCharges' togs bort var fÃ¶r att den innehÃ¥ller ungefÃ¤r samma information som tvÃ¥ andra kolumner, tenure och MonthlyCharges. Eftersom TotalCharges oftast bara Ã¤r hur lÃ¤nge nÃ¥gon varit kund ('tenure') multiplicerat med hur mycket de betalar varje mÃ¥nad ('MonthlyCharges'), kÃ¤nde jag att det inte gav nÃ¥gon nyttig extra information.

NÃ¤r det gÃ¤ller gender, sÃ¥g jag att om en kund var man eller kvinna inte verkade pÃ¥verka om de skulle sluta anvÃ¤nda tjÃ¤nsten eller inte. SÃ¥ jag bestÃ¤mde mig fÃ¶r att inte inkludera kundens kÃ¶n i modellen, eftersom det inte verkade hjÃ¤lpa till att fÃ¶rutsÃ¤ga kundavhopp. Genom att ta bort dessa tvÃ¥ kolumner fÃ¶rsÃ¶kte jag gÃ¶ra datan enklare och mer fokuserad fÃ¶r mina analyser.

```python
columns_to_drop = ['TotalCharges','gender']
train_df = train_df.drop(columns=columns_to_drop)
test_df = test_df.drop(columns=columns_to_drop)
```

#Det finns en hel del textdata som mÃ¥ste behandlas pÃ¥ nÃ¥got vis (finns flera olika alternativ, dÃ¤r vissa metoder Ã¤r mer lÃ¤mpade fÃ¶r neurala nÃ¤tverk)

Jag har valt att anvÃ¤nda OneHotEncoder fÃ¶r att hantera kategorisk textdata. Denna teknik omvandlar varje kategori till en binÃ¤r form och undviker att modellen felaktigt tolkar kategorier som ordnade (som kan hÃ¤nda om man tilldelar varje kategori ett heltal). PÃ¥ sÃ¥ sÃ¤tt behandlas varje kategori som unik och oberoende. FÃ¶r data med bara ja eller nej har jag valt att anvÃ¤nda en enklare binÃ¤r kodningsmetod. Denna metod mappar direkt "Ja" till 1 och "Nej" till 0. Det Ã¤r effektivt fÃ¶r dessa binÃ¤ra variabler eftersom det inte finns nÃ¥got behov av att skapa ytterligare kolumner, vilket bevarar datans ursprungliga struktur och minimerar komplexiteten. Dessutom sÃ¤kerstÃ¤ller denna teknik att modellen inte felaktigt tolkar binÃ¤ra vÃ¤rden som kontinuerliga variabler och bevarar deras distinkta natur.

```python
columns = ['Churn','PaperlessBilling','Dependents','Partner']

for col in columns:
    train_df[col] = train_df[col].map({'No': 0, 'Yes': 1})
    test_df[col] = test_df[col].map({'No': 0, 'Yes': 1})
```

```python
train_df = pd.get_dummies(train_df, drop_first=True)
test_df = pd.get_dummies(test_df, drop_first=True)

train_df, test_df = train_df.align(test_df, join='outer', axis=1, fill_value=0)
```

```python
X_train = train_df.drop('Churn', axis=1)
y_train = train_df['Churn']
X_test = test_df.drop('Churn', axis=1)
y_test = test_df['Churn']
```

# Klass Vikter

NÃ¤r vi trÃ¤nar en maskininlÃ¤rningsmodell fÃ¶r att fÃ¶rutsÃ¤ga nÃ¥got â€“ lÃ¥t oss sÃ¤ga om en kund kommer att avsluta en tjÃ¤nst (det vi kallar churn) â€“ stÃ¶ter vi ibland pÃ¥ ett problem dÃ¤r antalet kunder som stannar kvar Ã¤r mycket stÃ¶rre Ã¤n antalet som lÃ¤mnar. Detta gÃ¶r vÃ¥r data obalanserad eftersom det finns mycket mer information om den ena gruppen (de som stannar) Ã¤n den andra (de som lÃ¤mnar).
PÃ¥ sÃ¥ sÃ¤tt hjÃ¤lper klassvikter oss att bygga en rÃ¤ttvisare och mer balanserad modell som inte bara fokuserar pÃ¥ majoritetsgruppen, utan tar lika mycket hÃ¤nsyn till alla, oavsett hur mÃ¥nga de Ã¤r.

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights_dict = dict(enumerate(class_weights))
```

# TrÃ¤na olika neurala nÃ¤tverk

```python
configs = [
    (2, 512, 0.01),
    (3, 512, 0.001),
    (2, 1024, 0.001),
    (3, 1024, 0.0001),
    (2, 1024, 0.0005),
    (3, 768, 0.001),
    (4, 768, 0.0001),
    (2, 256, 0.02),
    (3, 256, 0.002),
    (2, 1536, 0.001),
    (3, 1536, 0.0001),
    (4, 640, 0.01),
    (2, 640, 0.001),
    (3, 640, 0.0001),
    (4, 576, 0.001),
    (2, 576, 0.0001),
    (3, 576, 0.00001),
    (4, 2048, 0.001),
    (2, 2048, 0.0001),
    (4, 2048, 0.001),
    (2, 2048, 0.0001),
    (3, 2048, 0.00001),
    (4, 1792, 0.0005),
    (2, 1792, 0.00005),
]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, Activation
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam, SGD
from sklearn.metrics import confusion_matrix
import time

def kerasClassifier(configs, X_train, y_train, X_test, y_test,class_weights):
    best_accuracy = 0
    best_model = None
    best_params = None
    for layers, neurons, learning_rate in configs:
        model = Sequential()
        model.add(Dense(neurons, input_dim=X_train.shape[1], activation='sigmoid'))
        model.add(Dropout(0.3))  # LÃ¤gg till ett Dropout-lager

        for _ in range(1, layers):
            model.add(Dense(neurons, activation='sigmoid'))
            model.add(Dropout(0.3))  # LÃ¤gg till Dropout-lager

        model.add(Dense(1,kernel_initializer='glorot_uniform', activation='sigmoid'))

        optimizer = Adam(learning_rate=learning_rate)
        model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        start_time = time.time()
        model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1, validation_data=(X_test, y_test), callbacks=[early_stopping], class_weight=class_weights)
        end_time = time.time()
        training_time = end_time - start_time

        _, accuracy = model.evaluate(X_test, y_test, verbose=1)
        y_pred = (model.predict(X_test) > 0.5).astype("int32")

        print(f"Neural network configuration: {layers} layers, {neurons} neurons per layer, learning rate {learning_rate}")
        print(f"Training time: {training_time:.2f} seconds")
        print(f"Accuracy: {accuracy:.2f}")

        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt="d")
        plt.title("Confusion matrix")
        plt.ylabel("Actual label")
        plt.xlabel("Predicted label")
        plt.show()

        if accuracy > best_accuracy:
          best_accuracy = accuracy
          best_model = model
          best_params = (layers, neurons, learning_rate)
          best_conf_matrix = cm

    return best_accuracy, best_model, best_params, best_conf_matrix

best_acccuracy, best_model, best_params, best_conf_matrix = kerasClassifier(configs, X_train, y_train, X_test, y_test,class_weights=class_weights_dict)
```

```python
print(f"Neural network configuration: {best_params[0]} layers, {best_params[1]} neurons per layer, learning rate {best_params[2]}\nAccuracy: {best_acccuracy:.2f}")
sns.heatmap(best_conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion matrix")
plt.ylabel("Actual label")
plt.xlabel("Predicted label")
plt.show()
```
